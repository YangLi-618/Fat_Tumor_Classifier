# Loading libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Loading the data

```python
# loading mimics datasets
from load_data_preprocess.load_data import reshape_data
Fat = reshape_data('Data/Trainging Data/20%Fat/')
Tumor = reshape_data('Data/Trainging Data/Tumor/')
data = pd.concat([pd.DataFrame(Fat), pd.DataFrame(Tumor)])
data.head()
```

# scaling the data before PCA

```python
# Using same normalization ways like we did before (are normalization)
from basic.basic import area_normalization
data = pd.DataFrame(area_normalization(data))
# When trying different dimensiona number, using data_new
data.head()
```

# implementing PCA

```python
# First let's reduce to 3-D
from sklearn.decomposition import PCA
pca = PCA(n_components=3).fit(data)
pca_samples = pca.transform(data)
```

```python
# We can see which features are important to the PC1 and PC2...
def pca_results(data, pca):
    # Dimension indexing
    dimensions = ['Dimension {}'.format(i) for i in range(1, len(pca.components_) + 1)]

    # PCA components
    components = pd.DataFrame(np.round(pca.components_, 4), columns=data.keys())
    components.index = dimensions

    # PCA explained variance
    ratios = pca.explained_variance_ratio_.reshape(len(pca.components_), 1)
    variance_ratios = pd.DataFrame(np.round(ratios, 4), columns=['Explained Variance'])
    variance_ratios.index = dimensions

    # Create a bar plot visualization
    fig, ax = plt.subplots(figsize=(20, 20))

    # Plot the feature weights as a function of the components
    components.plot(ax=ax, kind='bar')
    ax.set_ylabel("Feature Weights")
    ax.set_xticklabels(dimensions, rotation=0)

    # Display the explained variance ratios#
    for i, ev in enumerate(pca.explained_variance_ratio_):
        ax.text(i - 0.40, ax.get_ylim()[1] + 0.05, "Explained Variance\n %.4f" % (ev))

    # Return a concatenated DataFrame
    return pd.concat([variance_ratios, components], axis=1)
pca_results = pca_results(data, pca)
```

```python
# Showing the cumulative variance explained by the PCA in the above dataset
# From the above table, we can see that first 5 components together explains 95.92% variance in the data. 
pca_results.cumsum()
```

# Explained variance

```python
# Explained variance
# This plot shows cumulative explained variance on y-axis and number of principal components on the x-axis. 
# The number of principal components where we get the elbow is the optimal number of components that we should select for our analysis. 
plt.rcParams['figure.figsize'] = (8,5)
plt.plot(pca.explained_variance_ratio_)
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```
